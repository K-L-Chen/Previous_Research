```python
#nxn matrix,
#-1 means unreachable, 0 is self, posint is weight/distance
#indexing:
#   A B C D E
# A
# B
# C
# D
# E

graph_am = [
[ 0 , 1 , 5 ,-1 , -1],
[ 1 , 0 , 2 , 9 , -1],
[ 5 , 2 , 0 ,-1 , 3],
[-1 , 9 , -1, 0 , 2],
[-1 ,-1 , 3 , 2 , 0]
]

def shortest_distance(matrix:list, locationFrom=0):
    #current shortest length of path to location
    #A:i=0, B:i=1, C:i=2, D:i=3, E:i=4
    cur_shortest = list(matrix[locationFrom])

    for i in range(len(matrix)):
        for j in range(len(matrix)):
            print(cur_shortest)
            if i !=  locationFrom:# and j != locationFrom and i != j:
                if (cur_shortest[i] > 0 and matrix[i][j] > 0) and (matrix[i][j] + cur_shortest[i] < cur_shortest[j] or cur_shortest[j] < 0):
                    cur_shortest[j] = matrix[i][j] + cur_shortest[i]
                    #print(f"*******{cur_shortest[i]} {i} {j} {matrix[i][j]}*******")
                elif cur_shortest[j] < 0:
                    cur_shortest[j] = matrix[i][j]
                    #print(f"*******{i} {j} {matrix[i][j]}*******")
    
    return cur_shortest

def get_shortest_path(matrix:list, shortest_weights:list):
    path = []
    for nodeA in range(len(matrix)):
        for nodeB in range(len(matrix)):
            if nodeA != nodeB and (nodeA,nodeB) not in path:
                if matrix[nodeA][nodeB] + shortest_weights[nodeA] == shortest_weights[nodeB]:
                    path.append((nodeA,nodeB))
                    break
    return path
```
